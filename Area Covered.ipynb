```python
import matplotlib.image as img
import matplotlib.pyplot as plt
import numpy as np
import imageio

pic = imageio.imread('Clustering_3.png')
red_mask = pic[:, :, 0] < 252

pic[red_mask] = 0
plt.imshow(pic)
red = 0
total = 0
print(np.shape(pic))

for i in range (0,692):
    for j in range (0,978):
        if pic[i,j,0] == 252: 
            red += 1
            total += 1
        else: 
            total += 1
print(red/total)
    
```
